### 1. Gradient of the Sigmoid Function

The sigmoid function is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

We define:

$$
u(x) = 1 + e^{-x}, \quad \sigma(x) = u(x)^{-1}
$$

Then:

$$
u'(x) = \frac{d}{dx}(1 + e^{-x}) = -e^{-x}
$$

Using the chain rule:

$$
\frac{d\sigma(x)}{dx}
= \frac{d}{dx}\bigl(u(x)^{-1}\bigr)
= -u(x)^{-2} \cdot u'(x)
= - (1 + e^{-x})^{-2} \cdot (-e^{-x})
= \frac{e^{-x}}{(1 + e^{-x})^2}
$$

Now we express the result in terms of $\sigma(x)$:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
\Rightarrow
1 - \sigma(x) = \frac{e^{-x}}{1 + e^{-x}}
$$

Therefore:

$$
\sigma(x)\bigl(1 - \sigma(x)\bigr)
= \frac{1}{1 + e^{-x}} \cdot \frac{e^{-x}}{1 + e^{-x}}
= \frac{e^{-x}}{(1 + e^{-x})^2}
$$

So we get:

$$
\frac{d\sigma(x)}{dx} = \sigma(x)\bigl(1 - \sigma(x)\bigr)
$$

### 2. Gradient of the Tanh Function

The hyperbolic tangent is defined as:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Let:

$$
a(x) = e^x - e^{-x}, \quad b(x) = e^x + e^{-x}
$$

Then:

$$
\tanh(x) = \frac{a(x)}{b(x)}, \quad
a'(x) = e^x + e^{-x}, \quad
b'(x) = e^x - e^{-x}
$$

Using the quotient rule:

$$
\frac{d}{dx}\tanh(x)
= \frac{a'(x)b(x) - a(x)b'(x)}{b(x)^2}
$$

Substituting:

$$
\frac{d}{dx}\tanh(x)
= \frac{(e^x + e^{-x})(e^x + e^{-x}) - (e^x - e^{-x})(e^x - e^{-x})}{(e^x + e^{-x})^2}
$$

We compute:

$$
(e^x + e^{-x})^2 = e^{2x} + 2 + e^{-2x}, \quad
(e^x - e^{-x})^2 = e^{2x} - 2 + e^{-2x}
$$

Therefore:

$$
\frac{d}{dx}\tanh(x)
= \frac{\bigl(e^{2x} + 2 + e^{-2x}\bigr) - \bigl(e^{2x} - 2 + e^{-2x}\bigr)}{(e^x + e^{-x})^2}
= \frac{4}{(e^x + e^{-x})^2}
$$

Now, note that:

$$
\tanh^2(x)
= \left(\frac{e^x - e^{-x}}{e^x + e^{-x}}\right)^2
= \frac{(e^x - e^{-x})^2}{(e^x + e^{-x})^2}
$$

Thus:

$$
1 - \tanh^2(x)
= \frac{(e^x + e^{-x})^2 - (e^x - e^{-x})^2}{(e^x + e^{-x})^2}
= \frac{4}{(e^x + e^{-x})^2}
$$

So we conclude:

$$
\frac{d}{dx}\tanh(x) = 1 - \tanh^2(x)
$$
